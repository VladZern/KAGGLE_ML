```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns; seaborn.set()
```

```python
text1 = 'sea born futher lol'
text2 = 'and or are '
text3 = 'wer per are'
```

```python
data = pd.DataFrame([text1,text2, text3])
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
TF_IDF = TfidfVectorizer()
X_transform = TF_IDF.fit_transform(data[data.columns[0]]) # X - тексты
TF_IDF.get_feature_names() #дает закодированные слова
```

```python
pd.DataFrame(X_transform.toarray())
```

```python
X_transform.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_transform = vectorizer.fit_transform(data[data.columns[0]])
print vectorizer.get_feature_names()
print pd.DataFrame(X_transform.toarray())
print X_transform.shape
```
